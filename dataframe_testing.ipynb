```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('EY-data-science-challenge-pack/test.csv')

print(df)
```

## TimeSeries Experimentation for FRANCE

```python
FRA = df.loc[df['LOCATION'] == 'FRA']
#print(france)
```

### Tertiary Education Timeseries

```python
fig, ax = plt.subplots(1,1)
gdp = ax.plot(FRA['TIME'],FRA['EDU_TERTIARY_PC_WKGPOP'],'b')
ax2 = ax.twinx()
gni = ax2.plot(FRA['TIME'],FRA['TER_Enrolment_rate'],'g')

plt.plot()
plt.legend()
```

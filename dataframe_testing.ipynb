## Machine Learning Time

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score

#df_ML = df_male
#X = df_ML.drop(['LOCATION','id','TIME','GENDER','Quality_of_life_measure'],axis=1)
df_ML = df_male[['GDP_USD_CAP','Quality_of_life_measure']]
df_ML = df_ML.dropna(axis=0, how='any')
X = df_ML.drop(['Quality_of_life_measure'],axis=1).values.reshape(-1, 1)
Y = df_ML['Quality_of_life_measure'].values.reshape(-1, 1)
lm = linear_model.LinearRegression()
lm.fit(X,Y)

#pd.DataFrame(zip(X.columns, lm.coef_), columns = ['Features','Estimated Coeffs'])

fig, ax = plt.subplots(1,1,figsize=(18, 16))
ax.scatter(df_ML['GDP_USD_CAP'],Y)
plt.show()
```

Judging from the plot above it would be good to isolate the main cluster in the top left (i.e. get rid of some of the outliers). It's worth noticing that the gradients (ignoring the offset) all seem to have some correlation.

```python
df_ML = df_male[['GDP_USD_CAP','Quality_of_life_measure']]
df_ML_no_outliers = df_ML.loc[(df_ML['GDP_USD_CAP'] < 60000) & (df_ML['Quality_of_life_measure'] > 0.3)]
df_ML = df_ML_no_outliers.dropna(axis=0, how='any')
X = df_ML_no_outliers.drop(['Quality_of_life_measure'],axis=1).values.reshape(-1, 1)
Y = df_ML_no_outliers['Quality_of_life_measure'].values.reshape(-1, 1)
lm = linear_model.LinearRegression()
lm.fit(X,Y)

#pd.DataFrame(zip(X.columns, lm.coef_), columns = ['Features','Estimated Coeffs'])

fig, ax = plt.subplots(1,1,figsize=(18, 16))
ax.scatter(df_ML_no_outliers['GDP_USD_CAP'],Y)
ax.plot(X, lm.predict(X),color='r')
plt.show()
```

Rought GDP to QoL linear regression fit, does indicate a good level of correlation. Although I feel like I should definitely remove the offsets from each country and just determine the best average gradient fit.

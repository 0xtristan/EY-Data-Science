```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('EY-data-science-challenge-pack/train.csv')

df
```

```python
df_HDI = pd.read_csv('UN_DATA/HDI_data.csv',encoding='latin-1')
#countries = df_HDI['Unnamed:0']
df_HDI = df_HDI[[name for name in df_HDI.columns if 'Unnamed' not in name or 'Unnamed: 0' in name]]
df_HDI.rename(columns = {'Unnamed: 0':'LOCATION'}, inplace = True)
df_HDI = df_HDI.dropna(axis=0, how='any')
df_HDI
```

```python
#df_HDI.melt('country')
df_HDI['LOCATION']
df_HDI['LOCATION'] = df_HDI['LOCATION'].replace({' Hong Kong, China (SAR)':'Hong Kong'})
df_HDI
```

### Standardising country names to cross-match

```python
import country_converter as coco
#some_names = list(df_HDI['LOCATION'])
#standard_names = coco.convert(names=some_names, to='ISO3',not_found=None)
#print(standard_names)
#print(len(standard_names))
```

```python
for name in list(df_HDI['LOCATION']):
    df_HDI['LOCATION'] = df_HDI['LOCATION'].replace({name:coco.convert(name, to='ISO3',not_found=None)})
```

```python
df_HDI['LOCATION'] = df_HDI['LOCATION'].replace({' Ireland':'IRL'})
df_HDI['LOCATION'] = df_HDI['LOCATION'].replace({' Congo':'COG'})
df_HDI['LOCATION'] = df_HDI['LOCATION'].replace({' Samoa':'WSM'})
df_HDI
```

```python
df_HDI_unpivot = df_HDI.melt('LOCATION')
df_HDI_unpivot.rename(columns = {'variable':'TIME','value':'HDI'}, inplace = True)
df_HDI_unpivot.dtypes
df_HDI_unpivot['TIME'] = df_HDI_unpivot['TIME'].astype('int64')
df_HDI_unpivot
```

```python
df_merged = pd.merge(df, df_HDI_unpivot, how='left', on=['LOCATION', 'TIME'])
```

```python
df_merged['Quality_of_life_measure'].corr(df_merged['HDI'])
```

### BLI Data

```python
import os
os.listdir('./BLI')
```

```python
df_BLI = pd.DataFrame()
dflist = []
directory = os.path.join(os.getcwd(),"BLI")
for root,dirs,files in os.walk(directory):
    for f in files:
        if file.endswith('.csv'):
            df_temp = pd.read_csv('BLI/'+str(f))
            if "TIME" not in df_temp.columns:
                df_temp['TIME']=f.split('.')[0][3:]
            dflist.append(df_temp)
    df_BLI = pd.concat(dflist)
```

```python
df_BLI = df_BLI.drop('TIME.1',axis=1)
```

```python
df_BLI
```

```python
df_life_sat = df_BLI.loc[(df_BLI['INDICATOR']=='SW_LIFS') & ((df_BLI['Inequality']=='Men') | (df_BLI['Inequality']=='Women'))][['LOCATION','TIME','Inequality','Value']]
df_life_sat = df_life_sat.rename(columns={'Value':'LIFE_SAT','Inequality':'GENDER'})
df_life_sat['GENDER'] = df_life_sat['GENDER'].str.upper()
df_life_sat = df_life_sat.reset_index(drop=True)
```

```python
from sklearn import preprocessing
# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(df_life_sat['LIFE_SAT'].values.reshape(-1, 1))

# Run the normalizer on the dataframe
df_life_sat['LIFE_SAT_NORM'] = pd.DataFrame(x_scaled)
```

```python
df_life_sat
```

```python
df_merged = pd.merge(df_merged, df_life_sat, how='left', on=['LOCATION', 'TIME','GENDER'])
```

```python
df_merged.to_csv('ExtraCols.csv')
```

### Numbeo QoL index

```python
df_numbeo = pd.read_csv('numbeo.csv')
```

```python
df_numbeo
```

```python
df_numbeo_qol = df_numbeo[['Country','Period','Quality of Life Index']]
```

```python
df_numbeo_qol = df_numbeo_qol.loc[~df_numbeo_qol['Period'].str.contains('mid')]
```

```python
df_numbeo_qol['Period'] = df_numbeo_qol['Period'].replace({'2012-Q1':'2012','2013-Q1':'2013'})
```

```python
df_numbeo_qol = df_numbeo_qol.rename(columns={'Country':'LOCATION','Period':'TIME'})
df_numbeo_qol['TIME'] = df_numbeo_qol['TIME'].astype('int64')
```

```python
for name in list(df_numbeo_qol['LOCATION']):
    df_numbeo_qol['LOCATION'] = df_numbeo_qol['LOCATION'].replace({name:coco.convert(name, to='ISO3',not_found=None)})
df_numbeo_qol
```

```python
#df_numbeo_qol.drop('QoL_norm',axis=1)
df_numbeo_qol = df_numbeo_qol.reset_index(drop=True)
df_numbeo_qol['Numbeo_QoL_norm'] = pd.DataFrame(min_max_scaler.fit_transform(df_numbeo_qol['Quality of Life Index'].values.reshape(-1, 1)))
df_numbeo_qol.rename(columns = {'Quality of Life Index':'Numbeo_QoL'}, inplace = True)
df_numbeo_qol
```

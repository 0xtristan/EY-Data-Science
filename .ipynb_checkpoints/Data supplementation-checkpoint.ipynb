```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('EY-data-science-challenge-pack/train.csv')

df
```

```python
df_HDI = pd.read_csv('UN_DATA/HDI_data.csv',encoding='latin-1')
#countries = df_HDI['Unnamed:0']
df_HDI = df_HDI[[name for name in df_HDI.columns if 'Unnamed' not in name or 'Unnamed: 0' in name]]
df_HDI.rename(columns = {'Unnamed: 0':'country'}, inplace = True)
df_HDI = df_HDI.dropna(axis=0, how='any')
df_HDI
```

```python
#df_HDI.melt('country')
df_HDI['country']
df_HDI['country'] = df_HDI['country'].replace({' Hong Kong, China (SAR)':'Hong Kong'})
df_HDI
```

### Standardising country names to cross-match

```python
import country_converter as coco
some_names = list(df_HDI['country'])
standard_names = coco.convert(names=some_names, to='ISO3',not_found=None)
print(standard_names)
print(len(standard_names))
```

```python
for name in list(df_HDI['country']):
    df_HDI['country'] = df_HDI['country'].replace({name:coco.convert(name, to='ISO3',not_found=None)})
```

```python
df_HDI['country'] = df_HDI['country'].replace({' Ireland':'IRL'})
df_HDI['country'] = df_HDI['country'].replace({' Congo':'COG'})
df_HDI['country'] = df_HDI['country'].replace({' Samoa':'WSM'})
df_HDI
```

```python
df_HDI_unpivot = df_HDI.melt('country')
```

```python
pd.merge(df, df_HDI_unpivot, on='business_id', how='inner')
```
